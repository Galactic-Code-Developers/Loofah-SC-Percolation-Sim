# 03 — Transport and Global $I_c$ (Bottleneck Proxy)

```python
from loofah_sc_sim.geometry import generate_scaffold_graph
from loofah_sc_sim.coating import apply_coating_states
from loofah_sc_sim.transport import assign_local_ic, estimate_global_ic

G = generate_scaffold_graph(n_nodes=400, mean_degree=3.0, hierarchy_levels=3, seed=9)
H = apply_coating_states(G, continuity_p=0.7, defect_p=0.03, seed=9)
assign_local_ic(H, ic_scale=1.0, ts=1.0)

anchors_a = set(range(0,10))
anchors_b = set(range(390,400))

print('Estimated global Ic:', estimate_global_ic(H, anchors_a, anchors_b))
```
