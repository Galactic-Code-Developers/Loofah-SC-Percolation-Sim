# 01 — Scaffold Generation (Graph Proxy)

```python
import networkx as nx
import matplotlib.pyplot as plt
from loofah_sc_sim.geometry import generate_scaffold_graph

G = generate_scaffold_graph(n_nodes=300, mean_degree=3.0, hierarchy_levels=3, seed=42)
print('Nodes:', G.number_of_nodes())
print('Edges:', G.number_of_edges())

plt.figure(figsize=(6,6))
nx.draw(G, node_size=8, alpha=0.4)
plt.title('Loofah-derived scaffold (graph proxy)')
plt.show()
```
